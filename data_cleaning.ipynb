```python
pdf = tabula.read_pdf('60-Day-2-7.pdf', pages='all')
```

```python
type(pdf), len(pdf), type(pdf[0]) # one pandas df per page
```

```python
# most pages were properly loaded
pdf[0]
```

```python
# in some pages the first row becomes the column header
pdf[24]
```

```python
# get a copy of the correct column names
real_header = tuple(pdf[0].columns)
real_header
```

```python
# fix column names
for i, v in enumerate(pdf):
    if tuple(v.columns) != real_header:
        v = v.T.reset_index().T.reset_index(drop=True) # send column names back to the first row
        v.columns = real_header # rename column names
        pdf[i] = v # i, v are not real. only the list itself is real. overwrite the original df
```

```python
# verify all df's have the right column names
set([tuple(x.columns) for x in pdf])
```

```python
# merge all pages
df = pd.concat(pdf).reset_index(drop=True)
```

```python
df
```

```python
# replace weird charaters with spaces
df = df.replace(r'\r+|\n+|\t+', ' ', regex=True)
```

```python
df.to_csv('crime_data.csv')
```

#  Part 2

```python
# to_csv turns the index into a column. index_col=0 tells pandas that column is the index
df = pd.read_csv('crime_data.csv', index_col=0)
```

```python
df['Offense_Category'] = df.Offense.str.split(' - ').str[0]
df['Offense_Description'] = df.Offense.str.split(' - ').str[1]
df.Offense_Description = df.Offense_Description.str.replace("- ", " - ")
df.Offense_Description = df.Offense_Description.str.replace("Skateboar d", "Skateboard")
df.Offense_Description = df.Offense_Description.str.strip()
```

```python
df['Initial_Incident_Category'] = df['Initial Incident'].str.split(' -').str[0]
df['Initial_Incident_Description'] = df['Initial Incident'].str.split(' -').str[1:]
df.Initial_Incident_Description = df.Initial_Incident_Description.apply(lambda x: " ".join(x))
df.Initial_Incident_Description = df.Initial_Incident_Description.str.replace("- ", " - ")
df.Initial_Incident_Description = df.Initial_Incident_Description.str.replace("Skateboar d", "Skateboard")
df.Initial_Incident_Description = df.Initial_Incident_Description.str.strip()
```

```python
df['Final_Incident_Category'] = df['Final Incident'].str.split(' - ').str[0]
df['Final_Incident_Description'] = df['Final Incident'].str.split(' - ').str[1:]
df.Final_Incident_Description = df.Final_Incident_Description.apply(lambda x: " ".join(x))
df.Final_Incident_Description = df.Final_Incident_Description.str.replace("- ", " - ")
df.Final_Incident_Description = df.Final_Incident_Description.str.replace("Skateboar d", "Skateboard")
df.Final_Incident_Description = df.Final_Incident_Description.str.strip()
```

```python
df['Location_Type'] = df.Location.str.split(' - ').str[1:]
df['Location_Type'] = df['Location_Type'].apply(lambda x: "".join(x))
df['Location_Type'] = df['Location_Type'].str.replace("- ", "-")
df['Location_Type'] = df['Location_Type'].str.replace("On CampusResidential Facility", "On Campus Residential Facility")
df.Location = df.Location.str.split(' - ').str[0]
df.Location = df.Location.str.strip()
```

```python
df
```

```python
# drop raw columns
cols = list(df.columns)
df = df[cols[0:3] + cols[10:16] + cols[6:9] + cols[16:] + cols[9:10]]
df
```

```python
df['Date Reported']
```

```python
df = df.rename(columns={"Date Reported": "Date_Reported", "Date From": "Date_From", "Date To": "Date_To"})
```

```python
df.Date_Reported = df.Date_Reported.str.replace(r" - [A-Z]{3} at ", " ", regex=True)
df.Date_Reported = pd.to_datetime(df.Date_Reported, format="%m/%d/%y %H:%M")
df.Date_Reported = df.Date_Reported.dt.strftime('%Y-%m-%d %H:%M')
```

```python
df.Date_From = df.Date_From.str.replace(r" - [A-Z]{3} at ", " ", regex=True)
df.Date_From = df.Date_From.str.replace(r"\..*", "", regex=True)
df.Date_From = pd.to_datetime(df.Date_From, format="%m/%d/%y %H:%M")
df.Date_From = df.Date_From.dt.strftime('%Y-%m-%d %H:%M')
```

```python
df.Date_To = df.Date_To.str.replace(r" - [A-Z]{3} at ", " ", regex=True)
df.Date_To = df.Date_To.str.replace(r"\..*", "", regex=True)
df.Date_To = pd.to_datetime(df.Date_To, format="%m/%d/%y %H:%M")
df.Date_To = df.Date_To.dt.strftime('%Y-%m-%d %H:%M')
```
